```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)
plt.figure()
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#bar chart
categories=['A','B','C','D']
values=[20,35,30,25]
plt.figure()
plt.barh(categories,values)
plt.plot()
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("values")

#scatter plot
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.randn(100)
sizes=100*np.random.randn(100)
plt.figure()
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5)
plt.title("scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#pie chart
sizes=[30,20,25,15,10]
labels=['A','B','C','D','E']
plt.figure()
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("pie chart")

#show all
plt.show()
```

```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])

#basic operations
print ("array a:",a)
print ("array b:",b)
print("sum of array:",np.add(a,b))
print("difference:",np.subtract(a,b))
print("product:",np.multiply(a,b))
print("division:",np.divide(a,b))
print("square:",np.sqrt(a))
print("expo:",np.exp(a))

#aggregation op
print("minimum value a:",np.min(a))
print("maximum value a:",np.max(b))
print("mean value a:",np.mean(a))
print("standard value a:",np.std(b))
print("sum of all ele value a:",np.sum(a))

#reshaping array
c=np.array([[1,2],[3,4],[5,6]])
print("array c:")
print(c)
print("reshapd array c(2 rows,3 columns):")
print(np.reshape(c,(2,3)))

#transposing array
d=np.array([[1,2,3],[4,5,6]])
print("array d:")
print("transposed array d:")
print(np.transpose(d))







```
